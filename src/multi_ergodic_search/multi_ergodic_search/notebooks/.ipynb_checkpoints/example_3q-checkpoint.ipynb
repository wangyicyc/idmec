```python
import sys 
sys.path.append('..')
import jax
from functools import partial
from jax import grad, jacfwd, vmap, jit, hessian, value_and_grad
from jax.lax import scan
# from jax.ops import index_update, index
import jax.random as jnp_random
import jax.numpy as jnp
import numpy as np
from scipy.special import comb

import numpy as onp
# 动力学模型
from example_3q_lib.dynamics import DoubleIntegrator, SingleIntegrator, HomoDynamics

from example_3q_lib.ergodic_metric import ErgodicMetric # 计算遍历度指标
from example_3q_lib.obstacle import Obstacle  # 构建障碍物
from example_3q_lib.cbf import constr2CBF # 将碰撞约束转换为CBF
from example_3q_lib.fourier_utils import BasisFunc, get_phik, get_ck, get_ck_avg # 获取Fourier基函数和系数
from example_3q_lib.target_distribution import TargetDistribution # 构建目标分布
from example_3q_lib.cbf_utils import sdf2cbf # 将SDF转换为CBF
from example_3q_lib.connection_probability import pair_connection_doubleInt, gaussian_function # 将SDF转换为CBF
from IPython.display import clear_output
import matplotlib.pyplot as plt
# 增广拉格朗日法优化器
from example_3q_lib.mySolver import al_iLQR 
import yaml
import pickle as pkl
```

```python
_tsteps = 100
_robot_number = 4
# 初始化遍历性优化相关对象和参数
basis = BasisFunc(n_basis=[10,10])           # 创建二维傅里叶基函数（用于遍历性指标计算）
erg_metric = ErgodicMetric(basis)            # 遍历性评价指标Ergodic Metric
archetype_robot_model = DoubleIntegrator()             # 单积分器机器人模型
robot_model = HomoDynamics(robot_number=_robot_number, dynamics=archetype_robot_model)
target_distr = TargetDistribution()          # 目标分布对
workspace_bnds = jnp.array([[0.0, 1.0], [0.0, 1.0]])         # 工作空间边界,1x1米

# 读取障碍物配置文件
with open('../config/obs_3q.yml', 'r') as file:
    obs_info = yaml.safe_load(file)

obs = []                                     # 障碍物对象列表
cbf_constr = []                              # 控制屏障函数约束列表
for obs_dict in obs_info['obstacles']:       # 遍历所有障碍物配置
    _ob = Obstacle(obs_dict)                 # 创建障碍物对象
    obs.append(_ob)                          # 添加到障碍物列表
    cbf_constr.append(sdf2cbf(robot_model.dxdt, _ob.distance))  # 生成障碍物CBF约束

# 优化器参数设置
opt_args = {
    "workspace": workspace_bnds,
    "dt": 0.1,
    "tsteps": _tsteps,
    "R": np.diag([5e-3, 5e-3] * _robot_number),
    "Q_z": np.diag([1e-4, 1e-4, 1e-3, 1e-3] * _robot_number),
    "R_v": np.diag([5e-3, 5e-3] * _robot_number),
    "U_max": np.array([0.5, 0.5] * _robot_number),
    "U_min": np.array([-0.5, -0.5] * _robot_number),
    "x0" : jnp.array([0.1, 0.1, 0.1, 0.85, 0.85, 0.1, 0.85, 0.85]),               # 起点
    "xf" : jnp.array([0.18, 0.85, 0.85, 0.75, 0.15, 0.25, 0.85, 0.25]),              # 终点
    "alpha" : 0.85,
    "avoidance_radius" : 0.1,
    "robot_number" : _robot_number,
    "period_num" : 4,            # 轨迹周期数
    "com_radius": 0.5,
    "robot_pair" : [(0, 1),
                   (0, 2),
                   (0, 3),
                   (1, 2),
                   (1, 3),
                   (2, 3),],
    }

@vmap
def emap(x):
    """将状态归一化到工作空间坐标"""
    return jnp.array([
        (x[0]-workspace_bnds[0][0])/(workspace_bnds[0][1]-workspace_bnds[0][0]), 
        (x[1]-workspace_bnds[1][0])/(workspace_bnds[1][1]-workspace_bnds[1][0])
    ])

def barrier_cost(e):
    """边界屏障函数，防止轨迹超出工作空间"""
    return (jnp.maximum(0, e-1) + jnp.maximum(0, -e))**2

# 通用多机器人位置提取
def get_all_positions(x_traj, robot_num, nx=4):
    positions = []
    for i in range(robot_num):
        pos = x_traj[:, i*nx : i*nx+2]
        positions.append(pos)
    return jnp.stack(positions)  # 形状 (robot_num, tsteps, 2)
# 定义轨迹优化的损失函数
def loss_func(sol, args):
    x_traj = sol['x']         # 轨迹点序列
    u_traj = sol['u']         # 控制输入序列
    N = args['tsteps']      # 离散点数
    dt = args['dt']         # 每步时间间隔
    tf = dt * N             # 轨迹总时间
    _nx = robot_model.nx
    phik = get_phik(target_distr.evals, basis)     # 目标分布的傅里叶系数 
    # for i in range(opt_args["robot_number"]):
    #     ck += get_ck(x_traj[:, i * _nx : (i + 1) * _nx], basis, tf, dt)
    ck = get_ck_avg(x_traj, basis, tf, dt, _robot_number, _nx)   # 轨迹的傅里叶系数
    erg_met = 3 * erg_metric(ck, phik)
    bar_cost = 0.0
    for i in range(opt_args["robot_number"]):
        # 对每个机器人计算其位置映射到工作空间后的边界回避代价
        bar_cost += barrier_cost(emap(x_traj[:, i * _nx : i * _nx + 2]))
    # 计算控制输入代价（能量消耗），使用二次型代价函数
    ctrl_cost = jnp.sum(0.5 * args["R"] @ u_traj.T * u_traj.T) / args["tsteps"]
    """ Traj opt loss function, not the same as erg metric """
    # return jnp.sum(barrier_cost(e)) + erg_met  # 损失=轨迹边界惩罚+总时间（目标是轨迹短且不越界）
    return erg_met + ctrl_cost + bar_cost.sum()
```

```python
func_pair = partial(
    gaussian_function,
    sigma=opt_args["com_radius"],
)
func_pair_connection_probability = jit(
    partial(
        pair_connection_doubleInt,
        _func_pair=func_pair,
        robot_pair=opt_args["robot_pair"],
        _nx=robot_model.nx,
        period_num=opt_args["period_num"],
    )
)
```

```python
# 定义动力学等式约束
def eq_constr(sol, args):
    """ dynamic equality constriants """
    x = sol['x']         # 轨迹点序列
    u = sol['u']         # 控制输入序列

    x0 = args['x0']         # 起点
    xf = args['xf']         # 终点
    N = args['tsteps']           # 离散点数
    dt = args['dt']               # 每步时间间隔
    tf = dt * N       # 轨迹总时间
    # 1. 提取所有机器人位置
    robots_pos = get_all_positions(sol['x'], opt_args['robot_number'])
    
    # 2. 起点约束
    start_cons = jnp.concatenate([
        robots_pos[i, 0] - args['x0'][i*2:i*2+2] 
        for i in range(opt_args['robot_number'])
    ])
    
    # 3. 终点约束
    end_cons = jnp.concatenate([
        robots_pos[i, -1] - args['xf'][i*2:i*2+2]
        for i in range(opt_args['robot_number'])
    ])
    return jnp.vstack([
        start_cons, 
        # x[1:,:]-(x[:-1,:]+dt*vmap(robot_model.dxdt)(x[:-1,:], u[:-1,:])),  # 动力学约束
        end_cons
    ])

# 定义不等式约束（遍历性、控制、障碍物等）
def ineq_constr(sol, args):
    """ inequality constraints"""
    x_traj = sol['x']         # 轨迹点序列
    u_traj = sol['u']         # 控制输入序列
    N = args['tsteps']           # 离散点数
    dt = args['dt']               # 每步时间间隔
    tf = dt * N       # 轨迹总时间
    _nx = robot_model.nx
    _nu = robot_model.nu
    # 障碍物CBF约束
    _cbf_ineq = []
    for i in range(opt_args["robot_number"]):
        _cbf_ineq += [vmap(_cbf_ineq, in_axes=(0,0,None, None))(x_traj[:,i * _nx : (i + 1) * _nx], \
                     u_traj[:,i * _nu : (i + 1) * _nu], args['alpha'], dt).flatten() 
                   for _cbf_ineq in cbf_constr]
    # 机器人避障约束：确保机器人之间保持安全距离
    _avoidance_box = []
    # 遍历所有机器人对
    for i in range(opt_args["robot_number"] - 1):
        _xi = x_traj[:, i * _nx : i * _nx + 2]  # 提取第i个机器人的位置轨迹(x,y)
        for j in range(i + 1, opt_args["robot_number"]):
            _xj = x_traj[:, j * _nx : j * _nx + 2]  # 提取第j个机器人的位置轨迹(x,y)
            # 计算机器人之间距离约束：
            # 距离/avoidance_radius - 1 <= 0 等价于 距离 <= avoidance_radius
            _avoidance_box.append(
                (-jnp.linalg.norm((_xi - _xj), axis=1) / opt_args["avoidance_radius"] + 1)
            )
    # 将所有机器人对的避障约束展平，并归一化（除以机器人对数的组合数）
    _avoidance_arr = jnp.array(_avoidance_box).flatten() / comb(
        opt_args["robot_number"], 2
    )
    connection_probability = func_pair_connection_probability(traj=x_traj)
    # 控制输入约束
    _ctrl_box = [(jnp.abs(u_traj) - 1.).flatten()]
    return jnp.concatenate(_ctrl_box + _cbf_ineq + [_avoidance_arr] + [-connection_probability])
```

```python
# 双机器人初始化
start_pos = opt_args["x0"]
end_pos = opt_args["xf"]
_x_traj = jnp.linspace(
    start_pos,
    end_pos,
    opt_args["tsteps"] + 1,
)
# 速度导数计算 (前向差分)
_x_dot = jnp.vstack([
    ((_x_traj[1:, :] - _x_traj[:-1, :]) / opt_args["dt"]),  # 速度计算
    np.zeros(shape=(2 * opt_args["robot_number"])),  # 末端补零
])

_init_x_traj = jnp.column_stack(
    [_x_traj.reshape(-1, 2), _x_dot.reshape(-1, 2)]
).reshape(-1, _x_traj.shape[1] + _x_dot.shape[1])

_init_u_traj = np.zeros((opt_args["tsteps"], opt_args["robot_number"] * 2))

init_state = _init_x_traj[0, :]
init_sol = {"x": _init_x_traj[:-1], "u": _init_u_traj}  # 去除最后时间步
```

```python
# ====== 可视化 ====== #
# 创建绘图区域
fig, ax = plt.subplots(figsize=(10, 8))

# 机器人1轨迹 (绿色)
robot1_x = init_sol['x'][:, 0]  # 所有时间步的x位置
robot1_y = init_sol['x'][:, 1]  # 所有时间步的y位置
ax.plot(robot1_x, robot1_y, 'g-', linewidth=2, label='Robot 1 Trajectory')
ax.scatter(robot1_x, robot1_y, s=100, c='green', marker='.', label='Robot 1 Trajectory Point')
ax.scatter(robot1_x[0], robot1_y[0], s=100, c='green', marker='o', label='Robot 1 Start')
ax.scatter(robot1_x[-1], robot1_y[-1], s=150, c='green', marker='*', label='Robot 1 End')

# 机器人2轨迹 (蓝色)
robot2_x = init_sol['x'][:, 4]  # 索引4: 机器人2的x位置
robot2_y = init_sol['x'][:, 5]  # 索引5: 机器人2的y位置
ax.plot(robot2_x, robot2_y, 'b-', linewidth=2, label='Robot 2 Trajectory')
ax.scatter(robot2_x, robot2_y, s=100, c='blue', marker='.', label='Robot 2 Trajectory Point')
ax.scatter(robot2_x[0], robot2_y[0], s=100, c='blue', marker='o', label='Robot 2 Start')
ax.scatter(robot2_x[-1], robot2_y[-1], s=150, c='blue', marker='*', label='Robot 2 End')

robot3_x = init_sol['x'][:, 8]  # 索引4: 机器人2的x位置
robot3_y = init_sol['x'][:, 9]  # 索引5: 机器人2的y位置
ax.plot(robot3_x, robot3_y, 'b-', linewidth=2, label='Robot 2 Trajectory')
ax.scatter(robot3_x, robot3_y, s=100, c='black', marker='.', label='Robot 2 Trajectory Point')
ax.scatter(robot3_x[0], robot3_y[0], s=100, c='black', marker='o', label='Robot 2 Start')
ax.scatter(robot3_x[-1], robot3_y[-1], s=150, c='black', marker='*', label='Robot 2 End')
robot4_x = init_sol['x'][:, 12]  # 索引4: 机器人2的x位置
robot4_y = init_sol['x'][:, 13]  # 索引5: 机器人2的y位置
ax.plot(robot4_x, robot4_y, 'b-', linewidth=2, label='Robot 2 Trajectory')
ax.scatter(robot4_x, robot4_y, s=100, c='red', marker='.', label='Robot 2 Trajectory Point')
ax.scatter(robot4_x[0], robot4_y[0], s=100, c='red', marker='o', label='Robot 2 Start')
ax.scatter(robot4_x[-1], robot4_y[-1], s=150, c='red', marker='*', label='Robot 2 End')
# 添加工作空间边界 (文档4参数)
workspace = opt_args["workspace"]  # 假设已定义
ax.set_xlim(workspace[0][0], workspace[0][1])
ax.set_ylim(workspace[1][0], workspace[1][1])
ax.set_xlabel('X Position')
ax.set_ylabel('Y Position')
ax.set_title('Dual-Robot Initialization Trajectories')
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(loc='best')

plt.show()
```

```python
# 创建al_iLQR求解器实例，用于解决多机器人系统的优化控制问题
# 该求解器结合了增强拉格朗日方法和迭代线性二次调节器(iLQR)
solver = al_iLQR(
    args=opt_args,              # 优化问题的参数配置，包括时间步长、机器人数量等
    objective=loss_func,      # 目标函数，定义优化的代价函数
    dynamics=robot_model,  # 系统动力学模型，描述机器人运动规律
    inequality=ineq_constr,    # 不等式约束函数，定义系统需要满足的约束条件
    equality=eq_constr,
)
# 绘制目标分布
pdf_vals, grids = target_distr.evals
grids_x, grids_y = target_distr.get_grids()
init_dual=True
# # 调用求解器的solve方法，求解优化问题
# # 使用初始状态和初始解作为起点，进行最多2000次迭代优化
# sol = solver.solve(
#     x0=init_state,           # 系统的初始状态
#     init_sol=init_sol,       # 初始轨迹解（状态和控制输入轨迹）
#     max_iter=2000,          # 最大迭代次数
#     if_print=True           # 是否打印迭代过程中的优化信息
# )

# # 从求解结果中提取最优状态轨迹和控制输入轨迹
# x_traj = sol["x"]           # 最优状态轨迹
# u_traj = sol["u"]           # 最优控制输入轨迹
#                    # 显示当前迭代的可视化结
```

```python
# 迭代优化并动态可视化轨迹与障碍物分布
for kk in range(100):
    
    sol= solver.solve(
    x0=init_state,           # 系统的初始状态
    init_sol=init_sol,       # 初始轨迹解（状态和控制输入轨迹）
    init_dual=init_dual,
    max_iter=3,          # 最大迭代次数
    if_print=True           # 是否打印迭代过程中的优化信息
    )
    init_dual=False
    init_sol = sol
    X, Y = jnp.meshgrid(*[jnp.linspace(0,1)]*2)      # 生成工作空间网格
    pnts = jnp.vstack([X.ravel(), Y.ravel()]).T     # 展开为坐标点列表

    if kk % 1 == 0:                               # 每次都可视化
        clear_output(wait=True)                   # 清除上一次输出，动态刷新
        # 以下为障碍物绘制相关代码（可选）
        # for obs in traj_opt.obs:
        #     _patch = obs.draw()
        #     plt.gca().add_patch(_patch)
        plt.contourf(grids_x, grids_y, pdf_vals.reshape(grids_x.shape), levels=50, cmap='Blues', alpha=0.3)
        _mixed_vals = jnp.inf*jnp.ones_like(X)      # 初始化距离场
        for _obs in obs:                         # 遍历所有障碍物
            _vals = vmap(_obs.distance)(pnts).reshape(X.shape)  # 计算距离场
            _mixed_vals = jnp.minimum(_vals, _mixed_vals)        # 取最小距离
            plt.contour(X, Y, _vals.reshape(X.shape), levels=[-0.01,0.,0.01])  # 绘制障碍物边界
        
        trajectory = sol['x']

        plt.plot(start_pos[0], start_pos[1], 'g^', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(end_pos[0], end_pos[1], 'gs', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(start_pos[2], start_pos[3], 'r^', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(end_pos[2], end_pos[3], 'rs', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(start_pos[4], start_pos[5], 'b^', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(end_pos[4], end_pos[5], 'bs', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(start_pos[6], start_pos[7], 'k^', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        plt.plot(end_pos[6], end_pos[7], 'ks', 
                markersize=7, markerfacecolor='none', markeredgewidth=2)
        # plt.text(start_pos[0], start_pos[1], 'Start', fontsize=5, 
        #         verticalalignment='bottom', horizontalalignment='right',
        #         bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

        # # 添加终点文本标签
        # plt.text(end_pos[0], end_pos[1], 'End', fontsize=5, 
        #         verticalalignment='bottom', horizontalalignment='right',
        #         bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))
        # 绘制轨迹点和轨迹线
        plt.plot(trajectory[:,0], trajectory[:,1], 'g.', markersize=8)
        plt.plot(trajectory[:,0], trajectory[:,1], 'g-', linewidth=2)
        plt.plot(trajectory[:,4], trajectory[:,5], 'r.', markersize=8)
        plt.plot(trajectory[:,4], trajectory[:,5], 'r-', linewidth=2)
        plt.plot(trajectory[:,8], trajectory[:,9], 'b.', markersize=8)
        plt.plot(trajectory[:,8], trajectory[:,9], 'b-', linewidth=2)
        plt.plot(trajectory[:,12], trajectory[:,13], 'k.', markersize=8)
        plt.plot(trajectory[:,12], trajectory[:,13], 'k-', linewidth=2)

        # 设置图形属性
        plt.title('Final Optimized Trajectory')
        plt.xlabel('X Position')
        plt.ylabel('Y Position')
        plt.grid(True, alpha=0.3)
        plt.xlim(0, 1)
        plt.ylim(0, 1)

        # 添加图例
        # plt.legend(loc='best', framealpha=0.7)
        plt.tight_layout()

        plt.show()
```
