```python
robot_number = 4
# 初始化遍历性优化相关对象和参数
archetype_robot_model = DoubleIntegrator()             # 单积分器机器人模型
robot_model = HomoDynamics(robot_number=robot_number, dynamics=archetype_robot_model)
# 读取障碍物配置文件
# with open('../config/obs_3q.yml', 'r') as file:
#     obs_info = yaml.safe_load(file)

# obs = []                                     # 障碍物对象列表
# cbf_constr = []                              # 控制屏障函数约束列表
# for obs_dict in obs_info['obstacles']:       # 遍历所有障碍物配置
#     _ob = Obstacle(obs_dict)                 # 创建障碍物对象
#     obs.append(_ob)                          # 添加到障碍物列表
#     cbf_constr.append(sdf2cbf(robot_model.dxdt, _ob.distance))  # 生成障碍物CBF约束
```

```python
with open("../config/config.yaml", "r") as f:
    loaded = yaml.safe_load(f)
# 提取 opt_args 子字典
config = loaded["opt_args"]
workspace_size = config["workspace"]
# 按需还原为 array / diag 矩阵等
opt_args = {
    "workspace": workspace_size,
    "dt": config["dt"] + 0.010 * (workspace_size - 3.0),
    "tsteps": config["tsteps"],
    "R": np.diag(config["R"]),
    "Q_z": np.diag(config["Q_z"]),
    "R_v": np.diag(config["R_v"]),
    "U_max": np.array(config["U_max"]),
    "U_min": np.array(config["U_min"]),
    "V_max": np.array(config["V_max"]),
    "V_min": np.array(config["V_min"]),
    "x0": jnp.array(config["x0"]),
    "xf": jnp.array(config["xf"]),
    "alpha": config["alpha"],
    "avoidance_radius": config["avoidance_radius"],
    "weight_erg": config["weight_erg"],
    "robot_number": robot_number,
    "com_radius": config["com_radius"],
    "minimum_probability": config["minimum_probability"],
    "period_num": config["period_num"],
    "r_probability": config["r_probability"],
    "r_pointMatch": config["r_pointMatch"],
    "robot_pair": [tuple(pair) for pair in config["robot_pair"]],
}
tsteps = opt_args["tsteps"]
```

```python
raw = loaded["mapinfo_point"]
# 还原为 jnp.array
mapinfo_point = {
    "ids": raw["ids"],
    "timestamps": raw["timestamps"],
    "means": [jnp.array(m) for m in raw["means"]],
    "covs": [jnp.array(c) for c in raw["covs"]],
}
target_distr = TargetDistribution(workspace_size, len(mapinfo_point['ids']), mapinfo_point)          # 目标分布对
robot_distr = []
for robot_id in range(robot_number):
    map_copy = copy.deepcopy(mapinfo_point)
    robot_distr.append(TargetDistribution(workspace_size, 0, map_copy))
```

```python
# 更高效的方法 - 使用列表推导式
F_beta = jnp.full(tsteps, 1)
small_beta = jnp.full(tsteps, 1e-6)
n_small_betas = jnp.array([jnp.array([])] * robot_number)
F_beta_normalized = F_beta / jnp.sum(F_beta)  # 预先计算归一化值

# 使用列表推导式创建 betas
betas = [
    {
    'x': jnp.asarray(F_beta_normalized),
    'px': jnp.asarray(small_beta),
    'y': jnp.asarray(n_small_betas),
}
    for _ in range(robot_number)
]
```

```python
# 双机器人初始化
start_pos = opt_args["x0"]
end_pos = opt_args["xf"]

from four_robots_lib.tools import init_traj
_init_u_traj = np.zeros((tsteps, 2))
init_state = []
sol_trajs = []
for robot_id in range(robot_number):
    _init_x_traj, _init_state, init_fix_traj = init_traj(start_pos[robot_id*2: robot_id*2+2], end_pos[robot_id*2: robot_id*2+2],tsteps,opt_args["dt"])
    sol_trajs.append(
        {"x": _init_x_traj[:-1], 
         "u": _init_u_traj, 
         "px": init_fix_traj, 
         "y":[jnp.array([]) for _ in range(robot_number)]}
    )
    init_state.append(_init_state)
```

```python
# 假设你想创建包含 n 个 al_iLQR 结果的数组
n = robot_number  # 例如，创建 10 个结果
traj_solver = [al_iLQR(
    args=opt_args,
    objective=loss_traj,
    dynamics=robot_model,
    inequality=ineq_constr,
    equality=eq_constr,
    target_distr = robot_distr[_id].evals,
    robot_id = _id,
) for _id in range(n)]

for robot_id in range(n):
    sol_trajs[robot_id], conv = traj_solver[robot_id].solve(
        x0=init_state[robot_id],
        init_sol=sol_trajs[robot_id],
        beta=betas[robot_id],
        init_dual=True,
        max_iter=1,
        if_print=False,
    ) 
```

```python
decay_type = 'nodecay'
# decay_type = 'linear'
decay_type = 'type1'
```

```python
from four_robots_lib.plot_utils import plot_trajs, plot_line_graph
from four_robots_lib.tools import find_first_connected, exchange_info, init_traj, find_mapinfo_jax
from four_robots_lib.decay_utils import update_beta
from IPython.display import clear_output
itera = 10
kk = 0
# connection_threshold = 0.7 + (workspace_size - 3.0) * 0.1
# find_mapinfo_threshold = 0.7 + (workspace_size - 3.0) * 0.1
connection_threshold = 0.25 * workspace_size
find_mapinfo_threshold = 0.5 * workspace_size
init_dual = False
if decay_type == 'nodecay':
    save_path = '../figures/nodecay_exchangeInfo.png'
elif decay_type == 'linear':
    save_path = '../figures/linear_exchangeInfo.png'
elif decay_type == 'type1':
    save_path = '../figures/type1_exchangeInfo.png'
involved_robots = list(range(robot_number))
involved_robots = set(involved_robots)
flag = True
accumulated_time = 0
be_num = 1
object_value = []
local_metric = [
    {
        'time': [],
        'values': [],
    }
    for _ in range(robot_number)
]
global_metric = [
    {
        'time': [],
        'values': [],
    }
    for _ in range(robot_number)
]
# # 将 ids 转为 JAX 数组并排序
# ids = jnp.array(mapinfo_point['ids'])
# sort_idx = jnp.argsort(ids)   # 得到从小到大的索引顺序

# # 按排序索引重排 means
# point_set = jnp.vstack([mapinfo_point['means'][i] for i in sort_idx])
# 可以不排列，初始化的时候注意就行
point_set = jnp.vstack(mapinfo_point['means'])
_robot = -1 
last_exchange_time = {}
```

```python
def optimize_trajs(involved_robots, sol_trajs, betas, traj_solver, init_state, init_dual):
    to_remove = set()
    for robot_id in involved_robots:
        sol_trajs[robot_id], conv = traj_solver[robot_id].solve(
            x0=init_state[robot_id],
            init_sol=sol_trajs[robot_id],
            beta=betas[robot_id],
            init_dual=init_dual,
            max_iter=5,
            if_print=False,
        )
        if conv:
            to_remove.add(robot_id)
    return sol_trajs, to_remove

def update_accumulated_time(connect_timestep, accumulated_time, be_num):
    if connect_timestep is None:
        return 10 * be_num - accumulated_time, 10 * be_num
    accumulated_time += connect_timestep
    if accumulated_time >= 10 * be_num:
        connect_timestep -= accumulated_time - 10 * be_num
        accumulated_time = 10 * be_num
    return connect_timestep, accumulated_time

```

```python
batched_find_mapinfo = jax.vmap(
    find_mapinfo_jax,
    in_axes=(0, None, 0, None),  # (trajectory, point_set, valid_mask, threshold)
)
# 迭代优化并动态可视化轨迹与障碍物分布
while kk < itera:
    sol_trajs, to_remove = optimize_trajs(involved_robots, sol_trajs, betas, traj_solver, init_state, init_dual)
    involved_robots -= to_remove
    clear_output(wait=True)                   # 清除上一次输出，动态刷新
    init_dual = False
    
    logging.info(f"Complete the iteration {kk+1}")
    kk += 1    
    plot_trajs(start_pos, end_pos, sol_trajs, betas, False, robot_distr, save_path)    
    
    # if not involved_robots or kk == itera - 1:
    if True:
        for robot_id in range(robot_number):
            global_metric[robot_id]['time'].append(len(global_metric[robot_id]['time']))
            global_metric[robot_id]['values'].append(traj_solver[robot_id].return_erg_met(sol_trajs[robot_id], betas[robot_id], target_distr.evals))
        
        logging.info('converged')
        connect_timestep, robot_pair = find_first_connected(sol_trajs, connection_threshold, last_exchange_time)
        # gpt code
        # ✅ 向量化计算所有机器人的 encounter
        all_trajs = jnp.stack([sol_trajs[r]['x'] for r in range(robot_number)])
        valid_masks = jnp.stack([
            jnp.array([
                not robot_distr[r].check_peak_exists(p, i, 'explore')
                for i, p in enumerate(point_set)
            ])
            for r in range(robot_number)
        ])
        encounters = batched_find_mapinfo(all_trajs, point_set, valid_masks, find_mapinfo_threshold)

        logging.info(robot_distr[0]._dist_params['means'][0])
        # ✅ 从 batched 输出结构中直接取字段
        earliest_times = encounters['earliest_time']
        earliest_idx = jnp.argmin(earliest_times)
        # ✅ 提取对应最早事件信息
        encounter = {k: v[earliest_idx] for k, v in encounters.items()}
        # 判断是否存在有效相遇
        if encounter['time_step'] != -1:
            logging.info("prior exchange info")
            if connect_timestep is None or connect_timestep > int(encounter['time_step']):
                connect_timestep = int(encounter['time_step'])
                _robot = int(earliest_idx)
                robot_pair = []
                logging.info("no exchange info")
        # gpt code
        connect_timestep, accumulated_time = update_accumulated_time(connect_timestep, accumulated_time, be_num)
        if accumulated_time == 10 * be_num:
            if accumulated_time >= tsteps - 15:
                break
            logging.info("update map")
            robot_pair = []
            target_distr.update_map(accumulated_time, "reset", "read")
            point_set = jnp.vstack(target_distr._dist_params['means'])
            be_num += 1
            _robot = -1
        # 原本代码
        
        if _robot != -1:
            logging.info("find new map info")
            mean = target_distr._dist_params['means'][encounter['point_index']]
            cov = target_distr._dist_params['covs'][encounter['point_index']]
            get_info = {
                'ids': [encounter['point_index'],],
                'timestamps': [accumulated_time],
                'means': [mean,],
                'covs': [cov,],   
            }
            robot_distr[_robot].merge_map(get_info) 
            traj_solver[_robot].update_distribution(robot_distr[_robot].evals)
            involved_robots.add(_robot)
            _robot = -1
                
        sol_trajs, connected_pairs = exchange_info(sol_trajs, robot_pair, connect_timestep, robot_distr)
        betas = update_beta(sol_trajs, connected_pairs, betas, decay_type)
        if not involved_robots:
            involved_robots = connected_pairs
                
        if len(sol_trajs[0]['x']) > 0:
            init_state = [traj['x'][0, :] for traj in sol_trajs]     
            init_dual = True
            kk = 0
        else:
            break
        logging.info(f"connect time:{connect_timestep}, accumulated time:{accumulated_time}, and the robot pair:{robot_pair}")
```
