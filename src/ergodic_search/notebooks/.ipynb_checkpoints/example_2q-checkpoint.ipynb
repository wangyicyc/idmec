```python
import sys 
sys.path.append('..')
import jax
from functools import partial
from jax import lax
from jax import grad, jacfwd, vmap, jit, hessian, value_and_grad
from jax.lax import scan
# from jax.ops import index_update, index
import jax.random as jnp_random
import jax.numpy as jnp
import numpy as np
from jax.flatten_util import ravel_pytree

import numpy as onp
# 动力学模型
from single_robot_lib.dynamics import DoubleIntegrator, SingleIntegrator

from single_robot_lib.ergodic_metric import ErgodicMetric # 计算遍历度指标
from single_robot_lib.obstacle import Obstacle  # 构建障碍物
from single_robot_lib.cbf import constr2CBF # 将碰撞约束转换为CBF
from single_robot_lib.fourier_utils import BasisFunc, get_phik, get_ck # 获取Fourier基函数和系数
from single_robot_lib.target_distribution import TargetDistribution # 构建目标分布
from single_robot_lib.cbf_utils import sdf2cbf # 将SDF转换为CBF
from IPython.display import clear_output
import matplotlib.pyplot as plt
# 增广拉格朗日法优化器
from single_robot_lib.mySolver import al_iLQR 
import yaml
import pickle as pkl
import time
```

```python
# 初始化遍历性优化相关对象和参数
basis = BasisFunc(n_basis=[10,10])           # 创建二维傅里叶基函数（用于遍历性指标计算）
erg_metric = ErgodicMetric(basis)            # 遍历性评价指标Ergodic Metric
robot_model = DoubleIntegrator()             # 单积分器机器人模型
n, m = robot_model.nx, robot_model.nu        # 获取机器人状态和控制维度
target_distr = TargetDistribution()          # 目标分布对象
workspace_bnds = jnp.array([[0.0, 1.0], [0.0, 1.0]])         # 工作空间边界,1x1米

# 读取障碍物配置文件
with open('../config/obs_2q.yml', 'r') as file:
    obs_info = yaml.safe_load(file)

obs = []                                     # 障碍物对象列表
cbf_constr = []                              # 控制屏障函数约束列表
for obs_dict in obs_info['obstacles']:       # 遍历所有障碍物配置
    _ob = Obstacle(obs_dict)                 # 创建障碍物对象
    obs.append(_ob)                          # 添加到障碍物列表
    cbf_constr.append(sdf2cbf(robot_model.dxdt, _ob.distance))  # 生成障碍物CBF约束
# 优化器参数设置
opt_args = {
    "workspace": workspace_bnds,
    "dt": 0.1,
    "tsteps": 100,
    "R": np.diag([1e-3, 1e-3]),
    "Q_z": np.diag([1e-4, 1e-4, 1e-3, 1e-3]),
    "R_v": np.diag([5e-3, 5e-3]),
    "U_max": np.array([0.5, 0.5]),
    "U_min": np.array([-0.5, -0.5]),
    "x0" : jnp.array([0.1, 0.1]),               # 起点
    "xf" : jnp.array([0.90, 0.5]),              # 终点
    "alpha" : 0.85,
    "erg_met_coeff" : 7,
    "robot_number" : 1,
    }
@vmap
def emap(x):
    """将状态归一化到工作空间坐标"""
    return jnp.array([
        (x[0]-workspace_bnds[0][0])/(workspace_bnds[0][1]-workspace_bnds[0][0]), 
        (x[1]-workspace_bnds[1][0])/(workspace_bnds[1][1]-workspace_bnds[1][0])
    ])

def barrier_cost(e):
    """边界屏障函数，防止轨迹超出工作空间"""
    return (jnp.maximum(0, e-1) + jnp.maximum(0, -e))**2
```

```python
clip_step = 40
beta_regular = jnp.full(opt_args['tsteps'], 1)
beta_regular = beta_regular/jnp.sum(beta_regular)
beta_truncated = jnp.full(opt_args['tsteps'], 0)
beta_truncated = beta_truncated.at[clip_step:].set(2)
beta_truncated = beta_truncated/jnp.sum(beta_truncated)
 
beta_decayed = jnp.arange(opt_args['tsteps'])
beta_decayed = beta_decayed/jnp.sum(beta_decayed)
```

```python
def loss_regular(sol, args):
    x_traj = sol['x']         # 轨迹点序列
    u_traj = sol['u']         # 控制输入序列
    N = args['tsteps']      # 离散点数
    dt = args['dt']         # 每步时间间隔
    tf = dt * N             # 轨迹总时间
    e = emap(x_traj[:, 0:2])
    phik = get_phik(target_distr.evals, basis)     # 目标分布的傅里叶系数 
    ck = get_ck(e, basis, tf, dt, beta_regular)   # 轨迹的傅里叶系数
    erg_met = opt_args['erg_met_coeff'] * erg_metric(ck, phik)    
    # 计算障碍物回避代价，防止机器人离开工作空间
    _nx = robot_model.nx
    bar_cost = barrier_cost(emap(x_traj[:, 0:2]))
    ctrl_cost = jnp.sum(0.5 * args["R"] @ u_traj.T * u_traj.T) / args["tsteps"]
    """ Traj opt loss function, not the same as erg metric """
    return erg_met + ctrl_cost + bar_cost.sum()
def loss_truncated(sol, args):
    x_traj = sol['x']         # 轨迹点序列
    u_traj = sol['u']         # 控制输入序列
    N = args['tsteps']      # 离散点数
    dt = args['dt']         # 每步时间间隔
    tf = dt * N             # 轨迹总时间
    e = emap(x_traj[:, 0:2])
    phik = get_phik(target_distr.evals, basis)     # 目标分布的傅里叶系数 
    ck = get_ck(e, basis, tf, dt, beta_truncated)   # 轨迹的傅里叶系数
    erg_met = opt_args['erg_met_coeff'] * erg_metric(ck, phik)    
    # 计算障碍物回避代价，防止机器人离开工作空间
    _nx = robot_model.nx
    bar_cost = barrier_cost(emap(x_traj[:, 0:2]))
    ctrl_cost = jnp.sum(0.5 * args["R"] @ u_traj.T * u_traj.T) / args["tsteps"]
    """ Traj opt loss function, not the same as erg metric """
    return erg_met + ctrl_cost + bar_cost.sum()

def loss_decayed(sol, args):
    x_traj = sol['x']         # 轨迹点序列
    u_traj = sol['u']         # 控制输入序列
    N = args['tsteps']      # 离散点数
    dt = args['dt']         # 每步时间间隔
    tf = dt * N             # 轨迹总时间
    e = emap(x_traj[:, 0:2])
    phik = get_phik(target_distr.evals, basis)     # 目标分布的傅里叶系数 
    ck = get_ck(e, basis, tf, dt, beta_decayed)   # 轨迹的傅里叶系数
    erg_met = opt_args['erg_met_coeff'] * erg_metric(ck, phik)    
    # 计算障碍物回避代价，防止机器人离开工作空间
    _nx = robot_model.nx
    bar_cost = barrier_cost(emap(x_traj[:, 0:2]))
    ctrl_cost = jnp.sum(0.5 * args["R"] @ u_traj.T * u_traj.T) / args["tsteps"]
    """ Traj opt loss function, not the same as erg metric """
    return erg_met + ctrl_cost + bar_cost.sum()
```

```python
# 定义动力学等式约束
def eq_constr(sol, args):
    """ dynamic equality constriants """
    x = sol['x']         # 轨迹点序列
    u = sol['u']         # 控制输入序列

    x0 = args['x0']         # 起点
    xf = args['xf']         # 终点
    N = args['tsteps']           # 离散点数
    dt = args['dt']               # 每步时间间隔
    tf = dt * N       # 轨迹总时间
    return jnp.vstack([
        x[0, :2] - x0,          # 起点约束
        # x[1:,:]-(x[:-1,:]+dt*vmap(robot_model.dxdt)(x[:-1,:], u[:-1,:])),  # 动力学约束
        x[-1, :2] - xf          # 终点约束
    ])

# 定义不等式约束（遍历性、控制、障碍物等）
def ineq_constr(sol, args):
    """ inequality constraints"""
    x = sol['x']         # 轨迹点序列
    u = sol['u']         # 控制输入序列
    N = args['tsteps']           # 离散点数
    dt = args['dt']               # 每步时间间隔
    tf = dt * N       # 轨迹总时间
    e = emap(x)             # 轨迹点归一化到工作空间
    # 障碍物CBF约束
    _cbf_ineq = [vmap(_cbf_ineq, in_axes=(0,0,None, None))(x, u, args['alpha'], dt).flatten() 
               for _cbf_ineq in cbf_constr]
    _ctrl_box = [(jnp.abs(u) - 1.).flatten()]
    # 合并所有不等式约束
    # return jnp.concatenate(_erg_ineq + _ctrl_box + _cbf_ineq)
    return jnp.concatenate(_ctrl_box )#+ _cbf_ineq)
    # 下面为其他约束的可选写法（已注释）
    # return jnp.array([erg_metric(ck, phik) - 0.001, -tf] + [(jnp.abs(u) - 2.).flatten()])
    # return jnp.array(0.)
    # p = x[:,:2] # extract just the position component of the trajectory
    # # obs_val = [vmap(_ob.distance)(p).flatten() for _ob in self.obs]
    # obs_val = [vmap(_cbf_ineq)(x, u).flatten() for _cbf_ineq in self.cbf_consts]
    # ctrl_box = [(jnp.abs(u) - 2.).flatten()]
    # _ineq_list = ctrl_box +
```

```python
# 针对单个机器人的初始化逻辑
_start_pos = opt_args["x0"]
_end_pos = opt_args["xf"]
# 使用指定的起点和终点位置生成轨迹
_x_traj = jnp.linspace(
    _start_pos,
    _end_pos,
    opt_args["tsteps"] + 1,
)
_x_dot = jnp.vstack(
    [
        ((_x_traj[1:, :] - _x_traj[:-1, :]) / opt_args["dt"]),
        np.zeros(shape=(2,))
    ]
)
_init_x_traj = jnp.column_stack(
    [_x_traj.reshape(-1, 2), _x_dot.reshape(-1, 2)]
).reshape(-1, _x_traj.shape[1] + _x_dot.shape[1])
_init_u_traj = np.zeros((opt_args["tsteps"], 2))

init_state = _init_x_traj[0, :]
sol_regular = {"x": _init_x_traj[:-1], "u": _init_u_traj}
sol_truncated = {"x": _init_x_traj[:-1], "u": _init_u_traj}
sol_decayed = {"x": _init_x_traj[:-1], "u": _init_u_traj}
```

```python
# 创建al_iLQR求解器实例，用于解决多机器人系统的优化控制问题
# 该求解器结合了增强拉格朗日方法和迭代线性二次调节器(iLQR)
solver_regular = al_iLQR(
    args=opt_args,              # 优化问题的参数配置，包括时间步长、机器人数量等
    objective=loss_regular,      # 目标函数，定义优化的代价函数
    dynamics=robot_model,  # 系统动力学模型，描述机器人运动规律
    inequality=ineq_constr,    # 不等式约束函数，定义系统需要满足的约束条件
    equality=eq_constr,
)
solver_truncated = al_iLQR(
    args=opt_args,             
    objective=loss_truncated, 
    dynamics=robot_model,  
    inequality=ineq_constr, 
    equality=eq_constr,
)
solver_decayed = al_iLQR(
    args=opt_args,             
    objective=loss_decayed, 
    dynamics=robot_model,  
    inequality=ineq_constr, 
    equality=eq_constr,
)
```

```python
# 绘制目标分布
pdf_vals, grids = target_distr.evals
grids_x, grids_y = target_distr.get_grids()
init_dual = True
```

```python
decayed_alpha = jnp.clip(beta_decayed / beta_decayed.max(), 0.05, 1) 
trunted_alpha = jnp.clip(beta_truncated / beta_truncated.max(), 0.05, 1) 
regular_alpha = jnp.clip(beta_regular / beta_regular.max(), 0.05, 1)
```

```python
for kk in range(50):    
    sol_regular, finished_regular = solver_regular.solve(
        x0=init_state,           # 系统的初始状态
        init_sol=sol_regular,       # 初始轨迹解（状态和控制输入轨迹）
        init_dual=init_dual,     
        max_iter=3,          # 最大迭代次数
        if_print=False,           # 是否打印迭代过程中的优化信息
    )
    sol_truncated, finished_truncated = solver_truncated.solve(
        x0=init_state,           
        init_sol=sol_truncated,      
        init_dual=init_dual,     
        max_iter=3,   
        if_print=False, 
    )
    sol_decayed, finished_decayed = solver_decayed.solve(
        x0=init_state,    
        init_sol=sol_decayed,   
        init_dual=init_dual,     
        max_iter=3,  
        if_print=False, 
    )
    init_dual = False
    clear_output(wait=True)

    
    # 在循环外部创建图形和子图
    fig = plt.figure(figsize=(24, 6), facecolor='none')
    ax1 = fig.add_subplot(1, 3, 1)
    ax2 = fig.add_subplot(1, 3, 2)
    ax3 = fig.add_subplot(1, 3, 3)    
    # 设置子图属性（只需设置一次）
    for ax in [ax1, ax2, ax3]:
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)
    
    # 预先绘制静态背景元素（如果不变）
    # 注意：如果背景会变化，这部分需要移到循环内
    ax1.contourf(grids_x, grids_y, pdf_vals.reshape(grids_x.shape), levels=50, cmap='Blues', alpha=0.3)
    ax2.contourf(grids_x, grids_y, pdf_vals.reshape(grids_x.shape), levels=50, cmap='Blues', alpha=0.3)
    ax3.contourf(grids_x, grids_y, pdf_vals.reshape(grids_x.shape), levels=50, cmap='Blues', alpha=0.3)
    # 绘制障碍物（如果不变）
    _mixed_vals = jnp.inf*jnp.ones_like(X)
    # for _obs in obs:
    #     _vals = vmap(_obs.distance)(pnts).reshape(X.shape)
    #     _mixed_vals = jnp.minimum(_vals, _mixed_vals)
    #     ax1.contour(X, Y, _vals.reshape(X.shape), levels=[-0.01, 0., 0.01], colors='black', linewidths=1.2)
    #     ax2.contour(X, Y, _vals.reshape(X.shape), levels=[-0.01, 0., 0.01], colors='black', linewidths=1.2)
    
    # 绘制起点终点（如果不变）
    start_point = opt_args["x0"]
    end_point = opt_args["xf"]
    ax1.plot(start_point[0], start_point[1], 'b^', markersize=8, markerfacecolor='none', markeredgewidth=2)
    ax1.plot(end_point[0], end_point[1], 'bs', markersize=8, markerfacecolor='none', markeredgewidth=2)
    ax2.plot(start_point[0], start_point[1], 'b^', markersize=8, markerfacecolor='none', markeredgewidth=2)
    ax2.plot(end_point[0], end_point[1], 'bs', markersize=8, markerfacecolor='none', markeredgewidth=2)
    ax3.plot(start_point[0], start_point[1], 'b^', markersize=8, markerfacecolor='none', markeredgewidth=2)
    ax3.plot(end_point[0], end_point[1], 'bs', markersize=8, markerfacecolor='none', markeredgewidth=2)

    # 绘制新轨迹
    # ax1.plot(sol_regular['x'][:, 0], sol_regular['x'][:, 1], 'b-', linewidth=2, alpha=regular_alpha)
    # ax1.plot(sol_regular['x'][:, 0], sol_regular['x'][:, 1], 'b.', markersize=8, alpha=regular_alpha)
    # ax2.plot(sol_truncated['x'][clip_step:, 0], sol_truncated['x'][clip_step:, 1], 'b-', linewidth=2, alpha=trunted_alpha)
    # ax2.plot(sol_truncated['x'][:, 0], sol_truncated['x'][:, 1], 'b.', markersize=8, alpha=trunted_alpha)
    # ax3.plot(sol_decayed['x'][:, 0], sol_decayed['x'][:, 1], 'b-', linewidth=2, alpha=decayed_alpha)    
    # ax3.plot(sol_decayed['x'][:, 0], sol_decayed['x'][:, 1], 'b.', markersize=8, alpha=decayed_alpha)
    ax1.scatter(sol_regular['x'][:, 0], 
                      sol_regular['x'][:, 1], 
                      c='b',              # 颜色
                      s=10,                # 大小（对应 markersize）
                      alpha=regular_alpha, # 每个点的透明度
                      edgecolors='none')  # 可选：去除边缘
    ax2.scatter(sol_truncated['x'][:, 0], 
                      sol_truncated['x'][:, 1], 
                      c='b',              # 颜色
                      s=10,                # 大小（对应 markersize）
                      alpha=trunted_alpha, # 每个点的透明度
                      edgecolors='none')  # 可选：去除边缘
    ax3.scatter(sol_decayed['x'][:, 0], 
                      sol_decayed['x'][:, 1], 
                      c='b',              # 颜色
                      s=10,                # 大小（对应 markersize）
                      alpha=decayed_alpha, # 每个点的透明度
                      edgecolors='none')  # 可选：去除边缘
    # 更新标题
    ax1.set_title(f'a.Trajectory(regular)', 
                  fontsize=30, fontweight='bold')
    ax2.set_title(f'b.Trajectory(truncated)', 
                  fontsize=30, fontweight='bold')
    ax3.set_title(f'c.Trajectory(decayed)', 
                  fontsize=30, fontweight='bold')
    # 显示更新后的图形

    if finished_truncated and finished_regular and finished_decayed:
        plt.savefig('final_optimized_trajectory1.png', dpi=300, bbox_inches='tight')
        print("✅ 最终轨迹图像已保存！")
        ax1.clear()
        ax2.clear()
        ax3.clear()
        break
    else:
        display(fig)    
    ax1.clear()
    ax2.clear()
    ax3.clear()
```
